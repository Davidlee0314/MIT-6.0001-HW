```python
# Find prime factors of a number

from collections import Counter

def prime_test(num):
    factor = []
    
    for i in range(1, num + 1):
        if num % i == 0:
            factor.append(i)
    if len(factor) <= 2:
        return True
    else:
        return False
    
def get_factors(num):
    factor = []
    
    for i in range(1, num + 1):
        if num % i == 0:
            factor.append(i)
    return factor
        

def prime(num):
    prime_factor_num = {}
    
    if prime_test(num) == True:
        prime_factor_num.setdefault(num , 0)
        prime_factor_num[num] += 1
        return prime_factor_num
    else:
        factor = get_factors(num)
        num2 = int(num / factor[1])
        prime_factor_num.setdefault(factor[1] , 0)
        prime_factor_num[factor[1]] += 1
        a = prime(num2)
        prime_factor_num = dict(Counter(prime_factor_num) + Counter(a))
        return prime_factor_num 
    
```

```python
# Reduce all functions
# Define functions to find prime factors of a number
# Need to beware of instances when num < 0 or den < 0

class fraction():
    def __init__ (self, num, den):
        assert type(num) == int and type(den) == int and den != 0
        
        num_test = True
        den_test = True
        
        if num < 0:
            num = -num
            prime_factor_num = prime(num)
            num_test = False
        else:
            prime_factor_num = prime(num)
            
        if den < 0:
            den = -den
            prime_factor_den = prime(den)
            den_test = False
        else:
            prime_factor_den = prime(den)
    
        for key in prime_factor_num.keys():
            if key in prime_factor_den:
                while prime_factor_den[key] > 0 and prime_factor_num[key] > 0:
                    num = int(num / key)
                    den = int(den / key)
                    prime_factor_den[key] -= 1
                    prime_factor_num[key] -= 1
                    
        if (num_test == True and den_test == False) or (num_test == False and den_test == True):
            self.num = -num 
            self.den = den
        elif (num_test == False and den_test == False) or (num_test == True and den_test == True):
            self.num = num 
            self.den = den
         

    def __str__(self):
        if self.num == self.den or self.num == -self.den:
            return str(int(self.num / self.den))
        elif self.num % self.den == 0:
            return str(int(self.num / self.den))
        else:
            return str(self.num) + "/" + str(self.den)
        
    def float(self):
        return float(self.num / self.den)
    def __add__(self, other):
        num = self.num*other.den + self.den*other.num
        den = self.den*other.den
        return fraction(num, den)
    def __sub__(self, other):
        num = self.num*other.den - self.den*other.num
        den = self.den*other.den
        return fraction(num, den)
    def __mul__(self, other):
        num = self.num*other.num
        den = self.den*other.den
        return fraction(num, den)
    def __truediv__(self, other):
        num = self.num*other.den
        den = self.den*other.num
        return fraction(num, den)
    def __eq__(self, other):
        if self.num == other.num and self.den == other.den:
            return True
        else:
            return False   


#Test all functions
        
a = fraction(1500, 3000) + fraction(3000, 4000)
print("a =",a)
print("float a =",a.float())

b = fraction(100, 1500) - fraction(2000, 13000)
print("b =",b)

c = fraction(4500, 6500) * fraction(300, 700)
print("c =",c)

d = fraction(130, 169) / fraction(400, 600)
print("d =",d)

e = fraction(400, 600) == fraction(20, 30)
print("e =",e)

f = fraction(600, 800) + fraction(150, 600)
print("f =",f)

g = fraction(400, -500)
print("g =",g)

h = fraction(16, -2)
print("h =",h)

i = fraction(9, 3)
print("i =",i)
```
